```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
import warnings
warnings.filterwarnings('ignore')
```

Fields include

Rank - Ranking of overall sales

Name - The games name

Platform - Platform of the games release (i.e. PC,PS4, etc.)

Year - Year of the game's release

Genre - Genre of the game

Publisher - Publisher of the game

NA_Sales - Sales in North America (in millions)

EU_Sales - Sales in Europe (in millions)

JP_Sales - Sales in Japan (in millions)

Other_Sales - Sales in the rest of the world (in millions)

Global_Sales - Total worldwide sales.

```python
df_anos_recentes = df[df['Year']>=2013]
df_anos_recentes
```

```python
df_anos_recentes['Platform'].value_counts()
```

```python
xbox_ps = df_anos_recentes.query('Platform =="PS3" or Platform=="PS4" or Platform=="X360" or Platform=="XOne"')
xbox_ps 
```

```python
soma_vendas = xbox_ps.groupby('Platform')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].sum().reset_index()
soma_vendas
```

```python
type(soma_vendas)
```

```python
fig, ax = plt.subplots(2,2,figsize=(20,12))

sns.barplot(x='Platform', y=soma_vendas['NA_Sales'], data=soma_vendas,ax=ax[0,0],palette='crest')
ax[0,0].set_title('Total vendas jogos América do Norte')
ax[0,0].set_ylabel('Vendas_NA(milhões)')
ax[0,0].set_xlabel('Plataformas')

sns.barplot(x='Platform', y=soma_vendas['EU_Sales'], data=soma_vendas,ax=ax[1,0],palette='crest')
ax[1,0].set_title('Total vendas jogos Europa')
ax[1,0].set_ylabel('Vendas_EUR(milhões)')
ax[1,0].set_xlabel('Plataformas')

sns.barplot(x='Platform', y=soma_vendas['JP_Sales'], data=soma_vendas,ax=ax[0,1],palette='crest')
ax[0,1].set_title('Total vendas jogos Japão')
ax[0,1].set_ylabel('Vendas_JAP(milhões)')
ax[0,1].set_xlabel('Plataformas')

sns.barplot(x='Platform', y=soma_vendas['Other_Sales'], data=soma_vendas,ax=ax[1,1],palette='crest')
ax[1,1].set_title('Outras vendas jogos total')
ax[1,1].set_ylabel('Outras_vendas(milhões)')
ax[1,1].set_xlabel('Plataformas')

plt.show();
```

```python
sns.barplot(x='Platform', y=soma_vendas['Global_Sales'], data=soma_vendas,palette='rocket')
plt.title('Total vendas jogos globais')
plt.xlabel('Plataformas')
plt.ylabel('Vendas_Glo(milhões)')
plt.show()
```
